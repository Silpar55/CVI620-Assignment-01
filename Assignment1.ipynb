```python
import cv2
import numpy as np
import math
```

```python
# Part 1 

# background
img = np.ones((600, 600, 3), np.uint8) * 255

# circle 
radius = 30
thickness = 25

# colors
blue = (255, 0, 0) # BGR
green = (0, 255,0) # BGR
red = (0 ,0, 255)

# coordinates
top_coor = (300,250)
left_coor = (250,350)
right_coor = (350,350)

# circles 
cv2.circle(img, top_coor, radius, red, thickness) # top
cv2.circle(img, left_coor, radius, green, thickness) # left
cv2.circle(img, right_coor, radius, blue, thickness) # right


# Function to draw a white triangle cut that fully slices the ring
def draw_cut(center, angle_deg):
    cut_length = radius + thickness  # cover the whole thickness of the circle

    angle_rad = math.radians(angle_deg)
    angle_rad2 = angle_rad + math.radians(30)
    angle_rad3 = angle_rad - math.radians(30)

    # Two points forming the wide mouth of the triangle
    x1 = int(center[0] + cut_length * math.cos(angle_rad2))
    y1 = int(center[1] + cut_length * math.sin(angle_rad2))

    x2 = int(center[0] + cut_length * math.cos(angle_rad3))
    y2 = int(center[1] + cut_length * math.sin(angle_rad3))

    # Triangle: center and two outer points
    triangle = np.array([center, (x1, y1), (x2, y2)], np.int32)
    cv2.fillPoly(img, [triangle], (255, 255, 255))


# Add cuts for each circle, the angles will manage the direction of the triangle
draw_cut(top_coor, 90)   # facing downward
draw_cut(left_coor, 330)   # facing top-right
draw_cut(right_coor, 270) # facing top-left

# Text 
text = 'OpenCV'
coor = (215, 450)
font = cv2.FONT_HERSHEY_SIMPLEX
fontScale = 1.5
color = (0, 0, 0)
thickness = 5

cv2.putText(img, text, coor, font, fontScale, color, thickness)

# Display the image
cv2.imshow('Logo', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Part 2 

# img 1 
first_img = cv2.imread('./img/ayla-nova.jpeg')
# img 2 
second_img = cv2.imread('./img/kermit.jpeg')

# Resize any of the image to be able to do any operation 
height, width, _ = first_img.shape
second_img_resized = cv2.resize(second_img, (width, height))

alpha = 0.2

# blend = (1−α)⋅img1 + α⋅img2
blend = (1-alpha)*first_img + alpha*second_img_resized
# The result will be in float64, therefore we need to convert it into uint8
blend_uint8 = blend.astype(np.uint8)
# Write the blended img
blend_img = cv2.imwrite('./img/manual_blend.jpg', blend_uint8)


# Display the image
img = cv2.imread('./img/manual_blend.jpg')

cv2.imshow('manual blend', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
